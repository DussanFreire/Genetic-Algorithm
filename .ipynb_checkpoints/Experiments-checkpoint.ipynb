# Experiments

## 1 Apply the genetic algorithm to a population with the following characterists
* pc = 0.7
* pm = 0.001
* 20 genes
* 100 chromosomes
* 20 iterations

### All experiments results

```python
p = Population()
reports = p.run_experiment()
for report in reports:
    report.display_results()
```

```python
reports[0].get_results_in_dataframe()
```

### Best experiment result

```python
min_number_of_generation = min(list(map(lambda r: r.number_of_generations, reports)))
report_with_less_generations = list(filter(lambda r: r.number_of_generations ==min_number_of_generation,reports ))
for report in report_with_less_generations:
    report.display_results()  
```

### Worst experiment result

```python
max_number_of_generation = max(list(map(lambda r: r.number_of_generations, reports)))
report_with_more_generations = list(filter(lambda r: r.number_of_generations ==max_number_of_generation,reports ))
for report in report_with_more_generations:
    report.display_results()  
```
